### What we will learn:<br> - Effect of Y-scramling the data set<br>

#### Import all necessary stuff

```python
from rdkit import Chem
from sklearn.cross_validation import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
from sklearn.metrics import precision_score,recall_score
from copy import deepcopy
from sklearn.metrics import roc_curve, auc
from cohenskappa import cohens_kappa
import gzip
from pylab import *
```

#### Get the SD file with pre-calculated descriptors.

```python
inF = "./Patch2.sdf"
cpds = [x for x in  Chem.SDMolSupplier(inF) if x is not None]
```

```python
#inF_10mikroM = "./functionaldata.ready2model.sdf.gz.2class_hERGTL.CUT_1000nM.sdf.gz.removedSDtags.sdf.sdf.gz.descr.sdf.sdf.gz"
#cpds = [x for x in  Chem.ForwardSDMolSupplier(gzip.open(inF_10mikroM)) if x is not None]
```

#### Dump the descriptors into two lists (given the same ordering).

```python
observables = []
descriptors = []
for cpd in cpds:
    prop_name = cpd.GetPropNames()
    temp_list = []
    for property in prop_name:
        if property not in ['hERG_TL','value']:
            temp_list.append(float(cpd.GetProp(property)))
        elif property == 'hERG_TL':
            observables.append(int(cpd.GetProp(property)))
    descriptors.append(temp_list)
```

#### Convert the lists into numpy arrays.

```python
dataDescrs_array = np.asarray(descriptors)
dataActs_array   = np.array(observables)
```

#### Prepare empty lists to plot the performance of accuracy and kappa.

```python
acclist_realRF                          = []
acclist_realRF_predictTrain             = []
acclist_predictionOnTest_scrambledtrain  = []
acclist_predictionOnTrain_scrambledtrain = []
kappalist_realRF                           = []
kappalist_realRF_predictTrain              = []
kappalist_predictionOnTest_scrambledtrain  = []
kappalist_predictionOnTrain_scrambledtrain = []
```

#### Do the Y-scrambling. Loop over the actual learning for 30 times.

```python
for randomseedcounter in range(1,31):
    X_train,X_test,y_train,y_test = train_test_split(dataDescrs_array,dataActs_array,test_size=.4,random_state=randomseedcounter)
    y_train_scrambled = deepcopy(y_train)
    X_train_scrambled = deepcopy(X_train)
    np.random.shuffle(y_train_scrambled)
    np.random.shuffle(X_train_scrambled)
    #
    #
    # training was done on "scrambled" data - prediction on test set
    clf_RF_scrambled     = RandomForestClassifier()
    clf_RF_scrambled     = clf_RF_scrambled.fit(X_train,y_train_scrambled)
    y_predict_scrambled  = clf_RF_scrambled.predict(X_test)
    confmatrix_scrambled = metrics.confusion_matrix(y_test,y_predict_scrambled)
    kappa_pystatsmodels  = cohens_kappa(confmatrix_scrambled)['kappa']
    kappalist_predictionOnTest_scrambledtrain.append(kappa_pystatsmodels)
    acclist_predictionOnTest_scrambledtrain.append(clf_RF_scrambled.score(X_test,y_test))
    #
    #  training was done on "scrambled" data - prediction on train set
    y_predict_scrambled_predictTrain  = clf_RF_scrambled.predict(X_train)
    confmatrix_scrambled_predictTrain = metrics.confusion_matrix(y_train,y_predict_scrambled_predictTrain)
    kappa_pystatsmodels               = cohens_kappa(confmatrix_scrambled_predictTrain)['kappa']
    kappalist_predictionOnTrain_scrambledtrain.append(kappa_pystatsmodels)
    acclist_predictionOnTrain_scrambledtrain.append(clf_RF_scrambled.score(X_train,y_train))
```

#### Let's plot the performance of kappa.

```python
fig = plt.figure(figsize=(10,10))
fig.text(0.5,0.8,'kappa performance', horizontalalignment='center', fontsize=14)
ax3 = fig.add_subplot(221)
ax3.set_ylim([-0.4,1])
ax3.set_title("Scrambled TrainSet \n TestSet prediction")
ax3.plot(kappalist_predictionOnTest_scrambledtrain)
ax4 = fig.add_subplot(222)
ax4.set_title("Scrambled TrainSet \n TrainSet prediction")
ax4.plot(kappalist_predictionOnTrain_scrambledtrain)
ax4.set_ylim([-0.4,1])
#plt.savefig("scrambled_kappa_RF.png")
plt.show()
```

#### And now let's plot the performance of accuracy.

```python
fig = plt.figure(figsize=(10,10))
fig.text(0.5,0.8,'accuracy performance', horizontalalignment='center', fontsize=14)
ax3 = fig.add_subplot(221)
ax3.set_ylim([-0.4,1])
ax3.set_title("Scrambled TrainSet \n TestSet prediction")
ax3.plot(acclist_predictionOnTest_scrambledtrain)
ax4 = fig.add_subplot(222)
ax4.set_title("Scrambled TrainSet \n TrainSet prediction")
ax4.plot(acclist_predictionOnTrain_scrambledtrain)
ax4.set_ylim([-0.4,1])
#plt.savefig("scrambled_acc_RF.png")
plt.show()
```
